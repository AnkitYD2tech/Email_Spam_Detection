```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import streamlit as st
```

```python
df=pd.read_csv("spam.csv",encoding='latin1')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['v1']=df['v1'].replace(['ham','spam'],['not spam','spam'])
```

```python
df=df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
df
```

```python
x=df['v2']
y=df['v1']
```

```python
#train test split
(x_train,x_test,y_train,y_test)=train_test_split(y,x,test_size=0.2)
```

```python
cv=CountVectorizer(stop_words='english')
```

```python
features=cv.fit_transform(y_train)
```

```python
#creating model
model=MultinomialNB()
model.fit(features,x_train)
```

```python
#Test model
```

```python
features_test=cv.transform(y_test)
```

```python
print(model.score(features_test,x_test))
```

```python
message=cv.transform(["Embrace summer with 50% off!Whether you're coding from home or from a sunny beach, summer is the perfect time to level up. You still have time to grab 50% off Mimo Pro before the sale ends."]).toarray()
```

```python
result=model.predict(message)
```
